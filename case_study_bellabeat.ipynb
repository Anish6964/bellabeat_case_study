#### Define a Reusable Function for Correlation Plots
Defines a function `correlation_plot` that creates scatter plots to visualize the correlation between different numerical variables along with regression lines. This function will be used to explore relationships between activity metrics and calories burned.


```python
def correlation_plot(x, y, data, title, xlabel, ylabel):
    m, b = np.polyfit(data[x], data[y], 1)
    mean_y = data[y].mean()
    plt.scatter(data[x], data[y], alpha=0.7, c=data[y], cmap='coolwarm', label=f'Mean {ylabel}: {mean_y:.2f}')
    plt.plot(data[x], m*data[x] + b, '-', color='black')
    plt.colorbar(label=f'{ylabel} Distribution')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.axhline(mean_y, color='blue', linestyle='dashed', linewidth=1)
    plt.legend()
    plt.grid(True)
    plt.show()
```

#### Correlation Between Total Steps and Calories
Utilizes the `correlation_plot` function to display how total steps taken by users correlate with the calories they burn. This visualization helps in understanding how step count impacts calorie expenditure.


#### Correlation Between Total Distance and Calories
Analyzes the relationship between total distance traveled and calories burned, providing insights into the efficiency of different types of physical activities in terms of calorie expenditure.


#### Correlation Between Total Active Hours and Calories
Examines how the total hours spent in various activity levels correlate with the calories burned, highlighting the benefits of increased active hours on overall calorie burn.


```python
correlation_plot('TotalSteps', 'Calories', data, 'Correlation Between Total Steps and Calories Burned', 'Total Steps', 'Calories Burned')
correlation_plot('TotalDistance', 'Calories', data, 'Correlation Between Total Distance and Calories Burned', 'Total Distance', 'Calories Burned')
correlation_plot('TotalHours', 'Calories', data, 'Correlation Between Total Active Hours and Calories Burned', 'Total Active Hours', 'Calories Burned')
```

Key findings from the scatter plots above:
* There is a positive correlation as shown in a brown line in both graphs between the total steps and the calories burned and between total distance and calories burned.
* There are outliers in this data which can be caused by errors in data collection, miscalculations, data contamination, human error, etc.
* A weak positive correlation as shown in a brown line whereas the increase of the hours logged does not translate to more calories being burned. That is largely due to the average sedentary hours (green line) plotted at the 16 to 17 hours range.
* Some outliers in the data

#### Average Calories Burned by Day of the Week
Creates a bar graph to show the average number of calories burned by users on different days of the week. This can indicate which days are typically more active and could influence targeted marketing strategies.


```python
avg_calories_by_day = data.groupby('DayOfTheWeek')['Calories'].mean().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.figure(figsize=(10, 6))
avg_calories_by_day.plot(kind='bar', color='teal')
plt.title('Average Calories Burned by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Calories Burned')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

#### Activity Distribution Across the Week
Displays a pie chart representing the proportion of total activity recorded on each day of the week. This visualization helps identify days with higher or lower user engagement, which is crucial for planning weekly marketing activities.


```python
days_count = data['DayOfTheWeek'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.figure(figsize=(8, 8))
days_count.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired(range(len(days_count))))
plt.title('Distribution of Activity Days by Day of the Week')
plt.ylabel('')  # Hide the y-label as it is not informative for pie charts
plt.show()
```

#### Proportional Distribution of Activity Intensities
Shows a pie chart of the average minutes users spend in various activity intensities. This chart delineates the share of sedentary vs. active periods, offering strategic insights into user behavior patterns that Bellabeat can leverage in product positioning and marketing messages.


```python
# Aggregating minutes data
activity_minutes = data[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']].mean()

# Custom colors for the pie chart
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

# Creating the pie chart
plt.figure(figsize=(10, 8))
plt.pie(activity_minutes, labels=activity_minutes.index, autopct='%1.1f%%', startangle=140, colors=colors, explode=(0.1, 0, 0, 0))
plt.title('Percentage Distribution of Activity Minutes')
plt.show()
```

Key findings from the pie charts above:

The pie charts indicates that users were using the FitBit app to log for their daily activities.
Sedentary state of users make up the most (81.33%) of other activities in minutes.
The pie charts could indicate that the users did not use the app as intended such as working out, running, fitness, etc. Despite the fact that the app intention is to encourage fitness of the users.

# **Findings and Strategic Recommendations**
***

After thorough analysis and visualization of the FitBit usage data, we have extracted key behavioral insights that can be strategically applied to enhance Bellabeat's marketing approach. Below are the detailed findings and high-level recommendations based on the trends identified in the smart device usage data.

## **Key Findings:**
- **High Sedentary Activity Tracking**: A significant 81.3% of interactions with the app were used to track sedentary activities. This suggests that while users are interested in monitoring their health metrics, they may not be engaging deeply with the fitness potential of the devices.
- **Weekday Usage Predominance**: The data reveals that users are more active on the app during weekdays compared to weekends. This pattern indicates that user engagement aligns more with routine schedules, possibly linked to work or weekday activities.

## **Implications for Bellabeat Customers:**
- **Lifestyle Integration**: Bellabeat users, who are primarily women, likely follow similar patterns in utilizing smart devices to track health and fitness routines. The insights can thus be directly applied to enhance Bellabeat’s product offerings by integrating more lifestyle-oriented features.
- **Behavioral Insights**: The trend towards sedentary tracking provides Bellabeat an opportunity to encourage and promote more active and health-conscious behaviors through targeted features and campaigns.

## **Strategic Recommendations:**
- **Enhance Weekday Engagement**: Deploy motivational challenges and health tips during weekdays to increase user activity levels. For instance, introducing a 'Workweek Hustle' to encourage steps or active minutes could capitalize on the higher engagement during these days.
- **Stimulate Weekend Use**: Develop specific weekend campaigns that promote active use of the devices, such as weekend-only challenges or special rewards for meeting fitness goals, to counteract the drop in engagement.
- **Educational Campaigns**: Given the high usage for sedentary activity tracking, Bellabeat has an opportunity to educate its users on the benefits of an active lifestyle through in-app articles, notifications, and personalized insights.
- **Caloric Output Features**: Introduce or highlight features that allow users to track caloric intake and burn. Providing analytical tools that help users understand and manage their calorie consumption vs. expenditure could enhance the app's utility.
- **Smart Reminders**: Implement intelligent reminders that prompt users to move or exercise after periods of inactivity, especially tailored for users who primarily engage in sedentary activities.

By adopting these strategies, Bellabeat can not only increase engagement but also empower its users to lead healthier lives, leveraging both routine and spontaneous activities. This approach will help Bellabeat to strengthen its market position and align its product offerings more closely with consumer needs and behaviors.
